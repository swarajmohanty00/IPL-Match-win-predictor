```python
#batsman_runs--> Runs Scored by BatsMan
#total_runs-->Runs Scored
```

```python
total_score_df=(delivery_df.groupby(['match_id','inning','batting_team']).sum()['total_runs'].reset_index())
```

```python
total_score_df=total_score_df[total_score_df['inning']==1]
```

```python
match_df=(match_df.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id'))
```

```python
teams=list(match_df['team1'].unique())
teams
```

```python
teams=['Sunrisers Hyderabad',
 'Mumbai Indians',
 'Gujarat Titans',
 'Royal Challengers Bangalore',
 'Kolkata Knight Riders',
 'Kings XI Punjab',
 'Chennai Super Kings',
 'Rajasthan Royals', 
 'Delhi Capitals']
```

```python
#Replace Delhi Daredevils with Delhi capitals
match_df['team1']=match_df['team1'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['team2']=match_df['team2'].str.replace('Delhi Daredevils', 'Delhi Capitals')

match_df['team1']=match_df['team1'].str.replace('Gujarat Lions', 'Gujarat Titans')
match_df['team2']=match_df['team2'].str.replace('Gujarat Lions', 'Gujarat Titans')

match_df['team1']=match_df['team1'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match_df['team2']=match_df['team2'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
```

```python
match_df['team1'].unique()
```

```python
match_df=match_df[match_df['team1'].isin(teams)]
match_df=match_df[match_df['team2'].isin(teams)]
```

```python
match_df['team1'].unique()
```

```python
match_df.shape
```

```python
match_df.head()
```

```python
#df_applied is 0 means normal match and 1 means duckworth lewis rules are applied
match_df['dl_applied'].value_counts()
```

```python
#Removing dl_apllied = 1 records or rows
match_df = match_df[match_df['dl_applied']==0]
```

```python
match_df.shape
```

```python
match_df.columns
```

```python
match_df=match_df[['match_id', 'city','winner','total_runs']]
```

```python
delivery_df=delivery_df[delivery_df['inning']==2]
```

```python
delivery_df
```

```python
delivery_df.info()
```

```python
delivery_df=match_df.merge(delivery_df,on='match_id')
delivery_df
```

```python
delivery_df.columns
```

```python
delivery_df.rename(columns={ 'total_runs_x':'total_runs', 'total_runs_y':'Ball_score'},inplace=True)
```

```python
delivery_df
```

```python
delivery_df['Score']=delivery_df[['match_id','Ball_score']].groupby('match_id').cumsum()['Ball_score']
```

```python
delivery_df[['Ball_score','Score']]
```

```python
#New columns which has target
delivery_df['target_left']=(delivery_df['total_runs']+1)-delivery_df['Score']
```

```python
delivery_df
```

```python
#1 over 6 balls
#ipl has 20 over match 20*6 = 120 balls
delivery_df['Remaining Balls']=(120-((delivery_df['over']-1)*6 + delivery_df['ball']))
```

```python
delivery_df
```

```python
#How many wickets are left out
delivery_df['player_dismissed']=delivery_df['player_dismissed'].fillna('0')
```

```python
delivery_df
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].apply(lambda x:x if x == '0' else '1').astype('int64')
```

```python
delivery_df
```

```python
delivery_df[['match_id','player_dismissed']]
```

```python
delivery_df['Wickets']=delivery_df[['match_id','player_dismissed']].groupby('match_id').cumsum()['player_dismissed'].values
```

```python
delivery_df
```

```python
delivery_df['Wickets']=10 - delivery_df['Wickets']
```

```python
delivery_df
```

```python
#Current Run Rate (crr)
delivery_df['crr']=(delivery_df['Score'])*6/(120-delivery_df['Remaining Balls'])
```

```python
delivery_df
```

```python
#Required Run Rate(rrr)
delivery_df['rrr']=(delivery_df['target_left'])*6/(delivery_df['Remaining Balls'])
```

```python
delivery_df
```

```python
#Creating a column reult where 1 mean won 0 means lost
def result(row):
    if row['batting_team'] == row['winner']:
        return 1
    else:
        return 0
```

```python
delivery_df['result']=delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
delivery_df.columns
```

```python
model_df=delivery_df[['batting_team','bowling_team', 'city','Score', 'Wickets', 'Remaining Balls','target_left','crr', 'rrr', 'result']]
model_df
```

```python
model_df.isnull().sum()
```

```python
#Remove the records where city is missing
model_df[model_df['city'].isna()]
```

```python
model_df.dropna(inplace=True)
```

```python
model_df.isnull().sum()
```

```python
model_df.describe()
```

```python
model_df=model_df[model_df['Remaining Balls']!=0]
```

```python
model_df.describe()
```

```python
model_df=model_df.sample(model_df.shape[0])
model_df.sample()
```

```python
#Splitting the data into features and target
x = model_df.iloc[:,:-1]
y = model_df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
#splitting the data to train and test:
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# Label encoding
#One hot Encoding
#transform non numerical data to numerical using One hot Encoding

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf=ColumnTransformer([('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])],remainder='passthrough')
```

```python
trf
```

```python
#Building the model LogisticRegression
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(steps=[('step1',trf),('step2',LogisticRegression(solver='liblinear'))])
```

```python
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
y_pred
```

```python
#Accuracy Score
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
